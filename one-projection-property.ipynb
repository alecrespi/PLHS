```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random as r
```

```python
# HLS implementation

def randComposedInterval(*intervals):
    num = len(intervals)
    rand = r.randint(0, num-1)
    return np.random.uniform(intervals[rand][0], intervals[rand][1])

def randomPop(v: list):
    rindex = r.randint(0, len(v) - 1)
    ritem = v.pop(rindex)
    return v, ritem

def renderDimsPool(n, p):
    dims_pool = []
    for _ in range(p):
        dim = []
        for i in range(n):
            dim.append(i)
        dims_pool.append(dim)
    return dims_pool

def plotLHSSquared(n, p, LHS, grid=False):
    if p != 2:
        return
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.scatter(LHS[:, 0], LHS[:, 1], marker='o', c='b', s=1)
    if grid:
        for q in range(0, n):
            plt.axhline(y=q/n, color='g', linestyle='-.')
            plt.axvline(x=q/n, color='g', linestyle='-.')


# function that generates Latin Hypercubes 
def randLHS(n, p):
    LHS = np.zeros((n, p))
    dims_pool = renderDimsPool(n, p)
    for i in range(n):
        for k in range(p):
            dims_pool[k], q = randomPop(dims_pool[k])
            LHS[i, k] = np.random.uniform(q/n, (q+1)/n)
    return LHS

# randLHS(5, 2)

n, p = 10000, 2
LHS = randLHS(n, p)
plotLHSSquared(n, p, LHS, False)

# plot 3d just for fun
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.scatter(LHS[:, 0], LHS[:, 1], LHS[:, 2], c='r', marker='o')

```
